## 1. Comparision Operators : Used to compare two things

### Syntax :
```
Expression Operator Expression
```
`Expression can be any Column, Value, Function, Expression(Price * Quantity), SUBQUERY`

### EQUAL OPERATOR (=) : Checks two expresions are EQUAL

#### Retrieve all customers from USA

````python
%%sql
select *
from customers
where country = 'USA';
````

#### Retrieve all customers from Germany.

````python
%%sql
select *
from customers
where country = 'Germany';
````

### NOT EQUAL OPERATOR (!=, <>) : Checks two expresions are NOT EQUAL

#### Retrieve all the customers who are not from Germany

````python
%%sql
select *
from customers
where country != 'Germany';
````

### GREATOR THAN OPERATOR (>) : Checks two one expresions is GREATOR THAN another expression

#### Retrieve all customers with a score greator than 500

````python
%%sql
select *
from customers
where score > 500;
````

### GREATOR THAN EQUAL TO OPERATOR (>=) : Checks two one expresions is GREATOR THAN or EQUAL TO another expression

#### Retrieve all customers with score 500 or more

````python
%%sql
select *
from customers
where score >= 500;
````

### LESS THAN OPERATOR (<) : Checks two one expresions is LESS THAN another expression

#### Retrieve al customers with score less than 500

````python
%%sql
select *
from customers
where score < 500;
````

### LESS THAN EQUAL TO OPERATOR (<=) : Checks two one expresions is LESS THAN or EQUAL TO another expression

#### Retrieve all customers with a score 500 or less

````python
%%sql
select *
from customers
where score <= 500;
````

## 2. LOGICAL OPERATORS : Compare the conditions/expression and evalutes their truth vale as TRUE/FALSE

### AND Operator : All condition must be TRUE for this then only this will be TRUE

#### Retrieve all customers who are from USA and score > 500

````python
%%sql
select *
from customers
where country = 'USA'
    and score > 500;
````

### OR Operator : At least one of the condition must be TRUE then only this will be TRUE

#### Reterive all customers who are either from USA or they have score greator than 500.

````python
%%sql
select *
from customers
where country = 'USA'
    or score > 500;
````

### NOT Operator : (REVERSE) or Excludes the matching values/expression

#### Retrieve all customers who not belongs to USA

````python
%%sql
select *
from customers
where country != 'USA';
````

````python
%%sql
select *
from customers
where not country = 'USA';
````

#### Retrieve all customers with a score not less than 500

````python
%%sql
select *
from customers
where not score < 500;
````

````python
%%sql
select *
from customers
where score >= 500;
````

## 3. RANGE OPERATOR : Checks the given or specific Value lies between the given range or not

### BETWEEN : 
```
1. Used to check the value ies between range or not
2. Border's values are inclusive
```

![alt text](image.png)

#### Syntax:
```
SELECT * 
FROM tableName
WHERE columnName
BETWEEN lower_boundary and upper_boundary
```

#### Retrieve all cutomers whoes scores fals in the range between 100 and 500

##### Tradtional Method

````python
%%sql
select *
from customers
where score >= 100
    and score <= 500;
````

##### USING BETWEEN

````python
%%sql
select *
from customers
where score between 100 and 500;
````

## 4. MEMBERSHIP OPERATOR : Checks values in a data collection or not like (in list, etc..)

### IN OPERATOR : Checks that values that lies in data sequence

#### Syntax :
```
SELECT *
FROM tableName
WHERE columnName IN ('value1', 'value2',....);
```

#### Retrieve all customers who belongs from USA or Germany

##### Traditional Method:

````python
%%sql
select *
from customers
where country = 'USA'
    or country = 'Germany';
````

##### Using IN OPERATOR:

````python
%%sql
select *
from customers
where country in ('USA', 'Germany');
````

### NOT IN OPERATOR : Checks that values which don't lies in data sequence

#### Syntax :
```
SELECT *
FROM tableName
WHERE columnName NOT IN ('value1', 'value2',....);
```

#### Retrieve all customers who doesn't belongs from USA or Germany

##### Traditional Method : 

````python
%%sql
Select *
from customers
where country != 'USA'
    and country != 'Germany';
````

##### NOT IN Method:

````python
%%sql
select *
from customers
where country NOT IN ('USA', 'Germany');
````
