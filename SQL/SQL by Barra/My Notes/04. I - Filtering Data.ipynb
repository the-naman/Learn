# WHERE {operators}
```
1. Comparision : (=, (!=, <>), >, >=, <, <=)
2. Logical : (AND, OR, NOT)
3. Range : (BETWEEN)
4. Membership : (IN, NOT IN)
5. Search : (LIKE)

## 1. Comparision Operators : Used to compare two things

### Syntax :
```
Expression Operator Expression
```
`Expression can be any Column, Value, Function, Expression(Price * Quantity), SUBQUERY`

### EQUAL OPERATOR (=) : Checks two expresions are EQUAL

#### Retrieve all customers from USA

````python
%%sql
select *
from customers
where country = 'USA';
````

#### Retrieve all customers from Germany.

````python
%%sql
select *
from customers
where country = 'Germany';
````

### NOT EQUAL OPERATOR (!=, <>) : Checks two expresions are NOT EQUAL

#### Retrieve all the customers who are not from Germany

````python
%%sql
select *
from customers
where country != 'Germany';
````

### GREATOR THAN OPERATOR (>) : Checks two one expresions is GREATOR THAN another expression

#### Retrieve all customers with a score greator than 500

````python
%%sql
select *
from customers
where score > 500;
````

### GREATOR THAN EQUAL TO OPERATOR (>=) : Checks two one expresions is GREATOR THAN or EQUAL TO another expression

#### Retrieve all customers with score 500 or more

````python
%%sql
select *
from customers
where score >= 500;
````

### LESS THAN OPERATOR (<) : Checks two one expresions is LESS THAN another expression

#### Retrieve al customers with score less than 500

````python
%%sql
select *
from customers
where score < 500;
````

### LESS THAN EQUAL TO OPERATOR (<=) : Checks two one expresions is LESS THAN or EQUAL TO another expression

#### Retrieve all customers with a score 500 or less

````python
%%sql
select *
from customers
where score <= 500;
````
