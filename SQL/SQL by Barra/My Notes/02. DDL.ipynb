## ALL COMMANDS:
1. CREATE : to craete the tables
2. ALTER : to modify/update/change the existing table definition

### DDL : Used to create, define or modify the data in table in Database

#### CREATE : Used to create tables in Db

##### Syntax :
```
CREATE TABLE TableName(
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    CONSTRAINT constraint_if_any,
);
```

##### Task : Create a new table called Persons with columns (id, person_name, birth_date, and, phone)

````python
%%sql
create table persons (
    id INT NOT NULL,
    person_name varchar(50) NOT NULL,
    birth_date date,
    phone VARCHAR(15) NOT NULL,
    CONSTRAINT pk_persons PRIMARY KEY (id)
);
````

````python
%%sql
show tables;
````

##### Fetching newly created table records

````python
%%sql
select * from persons;
````

##### Saving DDL code as script

We can save the code as script to use it in future or to store the code
1. Go to db 
2. do a right click
3. click on `Script Table as`
4. Select as per code (CREATE, ALTER, DROP, SELECT, INSERT, UPDATE, DELETE, EXECUTE) as per need
5. Select `New Query Editor Window`
6. Then seen the code and save as you need
