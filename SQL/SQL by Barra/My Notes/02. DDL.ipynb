## ALL COMMANDS:
1. CREATE : to craete the tables
2. ALTER : to modify/update/change the existing table definition
3. DROP : to drop the whole table from the DB, it also delete the all records/data from that table

- *Note* : DESCRIBE : To see the SCHEMA of existing table

### DDL : Used to create, define or modify the data in table in Database

### CREATE : Used to create tables in Db

##### Syntax :
```
CREATE TABLE TableName(
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    ColumnName dataType(size_if_required) constraint_if_any,
    CONSTRAINT constraint_if_any,
);
```

##### Task : Create a new table called Persons with columns (id, person_name, birth_date, and, phone)

````python
%%sql
create table persons (
    id INT NOT NULL,
    person_name varchar(50) NOT NULL,
    birth_date date,
    phone VARCHAR(15) NOT NULL,
    CONSTRAINT pk_persons PRIMARY KEY (id)
);
````

````python
%%sql
show tables;
````

##### Fetching newly created table records

````python
%%sql
select * from persons;
````

### ALTER : Use to modify  / change / update the definition of existing table

##### Syntax :
1. ADD : 
`
ALTER TABLE table_name
ADD column_name datatype;
`

2. DROP : 
`
ALTER TABLE table_name
DROP COLUMN column_name;
`

3. RENAME COLUMN :
`
ALTER TABLE table_name
RENAME COLUMN old_name to new_name;
`

4. MODIFY :
`
ALTER TABLE table_name
MODIFY column_name new_datatype constraint;
`

5. ADD CONSTRAINT :
`
ALTER TABLE table_name
ADD CONSTRAINT constraint_name constraint_definition;
`

6. RENAME TABLE :
`
ALTER TABLE table_name
RENAME TO new_table_name;
`

##### DESCRIBE Command : To see the SCHEMA of the existing table

##### Add a new column Email in existing table persons
