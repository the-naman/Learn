### WHERE CLAUSE : To filter out the required DATA

##### Customers who have greater than 500 score

````python
%%sql
select * from customers where score > 500;
````

##### Customers who have greater score than 0

````python
%%sql
select * from customers where score > 0;
````

##### Customers who have not score = 0

````python
%%sql
select * from customers where score != 0;
````

##### Customers who belongs from Germany

````python
%%sql
select *
from customers
where country = 'Germany';
````

##### Name and Country of customers who belongs to Germany

````python
%%sql
select first_name,
    country
from customers
where country = 'Germany';
````

### ORDER BY COMMAND : To SORT the DATA

- ASC :  `Low --> High`
- DESC : `High --> Low`

Syntax : 
```
SELECT *
FROM table_Name
ORDER BY column_Name;
```
```
SELECT *
FROM table_Name
where {condition}
ORDER BY column_Name;
```

##### Retrieve all customers and sort the result by the Highest score first

````python
%%sql 
--Default
select *
FROM customers
ORDER BY score;
````

````python
%%sql
-- ASC
select *
FROM customers
ORDER BY score ASC;
````

````python
%%sql
-- DESC
select *
FROM customers
ORDER BY score DESC;
````

##### Retrieve all customers and sort the result by country

````python
%%sql
select *
from customers
order by country;
````

##### Retrieve all customers and sort the result by country and then again by the Highest score

````python
%%sql
select *
from customers
order by country ASC,
    score DESC;
````

##### Select the only Germany customers and list out from highest to lowest scorer

````python
%%sql
select *
from customers
where country = 'Germany'
order by score;
````

### GROUP BY : To aggregate the data : Combies the rows having same value

Syntax : 
```
SELECT * 
FROM table_Name
WHERE {condition}
GROUP BY {condition}
ORDER BY column_Name;
```

### Find out the toal Score by each country

````python
%%sql
select country as 'Country',
    sum(score) as 'Total Score'
from customers
group by country;
````

### Find out the highest Score by each country

````python
%%sql
select country as 'Country',
    MAX(score) as 'Highest Score'
from customers
group by country;
````

##### Find the total scores and total customers by each country

````python
%%sql
select country,
    count(first_name) as 'Total Persons',
    sum(score) as 'Total Score'
from customers
group by country;
````

##### Find the total scores and total customers by each country and sort countries by highest to lowest scores

````python
%%sql
select country,
    count(id) as 'Total Persons',
    sum(score) as 'Total_Score'
from customers
group by country
order by Total_Score desc;
````
