## ALL COMMANDS:
```
1. INSERT : to insert data into a table : (INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);)
2. INSERT using SELECT : to in_sert data in a table which data belongs to another table and we only select that much data which we need to insert. : (INSERT INTO target_table_name (col1, col2, col3, ....) SELECT col1, col2, col3, ... from source_table_name)
3. UPDATE : to update existing data within a table : (UPDATE table_name SET column1 = value1, column2 = value2 WHERE condition;)
4. DELETE : to delete records from a database table : (DELETE FROM table_name WHERE condition;)
```

### INSERT using SELECT

#### RULE:
```
1. Query the source table
2. make data/column valid for target table using column(for available column value)/ static value(for unavailable column value)
3. insert or copy the data from source table to target table
```

#### Copy the data from Customer table to Person table

````python
%%sql
select * from customers;
````

````python
%%sql
select * from persons;
````

````python
%%sql
## Rule 1: Select * from customers;
## Rule 2: SELECT id, first_name as full_name,  NULL as birth_date, 'unknown' as phone from customers;
## Rule 3: INSERT INTO PERSONS (id, full_name, birth_date, phone) 
#                                       SELECT id, first_name as full_name,  NULL as birth_date, 'unknown' as phone from customers;

insert into persons (id, full_name, birth_date, phone)
SELECT id,
    first_name as full_name,
    NULL as birth_date,
    'unknown' as phone
from customers;
````

````python
%%sql
select * from persons;
````

### UPDATE : To update the existing data in a table

#### Syntax :
```
UPDATE table_name
    SET column1 = value1,
        column2 = value2
WHERE {condition};
``` 

`Always use WHERE condtion otherwise it update all records of the table`

#### Change the score of the customer having id is 6 to 0.

````python
%%sql
SELECT * FROM customers;
````

````python
%%sql
UPDATE customers
SET score = 0
WHERE id = 6;
````

````python
%%sql
SELECT *
FROM customers;
````
