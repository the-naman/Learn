## ALL COMMANDS:
```
1. INSERT : to insert data into a table : (INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);)
2. UPDATE : to update existing data within a table : (UPDATE table_name SET column1 = value1, column2 = value2 WHERE condition;)
3. DELETE : to delete records from a database table : (DELETE FROM table_name WHERE condition;)
```

### *Description:*
#### DML : Data Manipulation Language : Used to manipulate the data of the table in DB

### INSERT : to insert / append values in the existing table

#### Syntax : 
```
INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);
```
